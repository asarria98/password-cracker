# Resultados MPI

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
palabras = ["hola", "clavo", "casero"]

for palabra in palabras:
    # Cargar los datos del archivo CSV sin encabezados
    df = pd.read_csv(f"results/{palabra}_tiempos_mpi.csv", header=None, names=['nucleos', 'tiempo'])

    # Crear el plot
    plt.plot(df.iloc[:, 0], df.iloc[:, 1], label=palabra)

    # Añadir etiquetas y título
    plt.xlabel('Número de Núcleos')
    plt.ylabel('Tiempo (s)')
    plt.title(f'Tiempos de ejecución - {palabra}')
    plt.legend()

    # Mostrar el plot
    plt.show()
```

Como ya vimos en la pruebas realizadas en el ejercicio de paralelización de openmp, a medida que incrementamos el número de núcleos, disminuye el tiempo de ejecución. Sin embargo también se obserba un aumento de tiempo al pasar de 4 a 6 núcleos. Esto probablemente se debe a que 4 es mulitplo de 8(bits) y 6 no, por lo tanto el ejercicio es más adecuado para esa cantidad de núcleos.
