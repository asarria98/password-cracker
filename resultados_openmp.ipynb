# Resultados Openmp

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
palabras = ["hola", "clavo", "casero"]

for palabra in palabras:
    # Cargar los datos del archivo CSV sin encabezados
    df = pd.read_csv(f"results/{palabra}_tiempos_openmp.csv", header=None, names=['nucleos', 'tiempo'])
    t1 = df.iloc[0]["tiempo"]
    list_mejora = [t1 / df.iloc[t]["tiempo"] for t in range(len(df))]

    # Crear el plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.plot(df.iloc[:, 0], df.iloc[:, 1], label=palabra)
    ax2.plot(df.iloc[:, 0], list_mejora, label="mejoras")

    # Añadir etiquetas y título
    plt.xlabel('Número de Núcleos')
    plt.ylabel('Tiempo (s)')
    plt.title(f'Tiempos de ejecución - {palabra}')
    plt.legend()

    # Mostrar el plot
    plt.show()
```

Como se puede observar, y como podíamos preveer a medida que incrementamos el número de núcleos, disminuye el tiempo de ejecución. Sin embargo también se obserba un aumento de tiempo al pasar de 4 a 6 núcleos. Esto probablemente se debe a que 4 es mulitplo de 8(bits) y 6 no, por lo tanto el ejercicio es más adecuado para esa cantidad de núcleos.
