# Resultados OPENMP

En este cuaderno analizaremos los resultados obtenidos en la pr√°ctica.

```python
import matplotlib.pyplot as plt

# Datos proporcionados
datos_hola = [(1, 2.92), (2, 1.89), (4, 1.90), (6, 1.41)]
datos_clavo = [(1, 11.55), (2, 9.31), (4, 9.07), (6, 9.45)]
datos_casero = [(1, 306.67), (2, 260.62), (4, 244.87), (6, 282.99)]

# Gr√°fico para "hola"
nucleos_hola = [dato[0] for dato in datos_hola]
tiempo_hola = [dato[1] for dato in datos_hola]

plt.figure(figsize=(8, 6))
plt.plot(nucleos_hola, tiempo_hola, marker='o', label='hola')
plt.title('Tiempo de ejecuci√≥n para "hola"')
plt.xlabel('N√∫mero de n√∫cleos')
plt.ylabel('Tiempo (segundos)')
plt.grid(True)
plt.legend()
plt.show()

# Gr√°fico para "clavo"
nucleos_clavo = [dato[0] for dato in datos_clavo]
tiempo_clavo = [dato[1] for dato in datos_clavo]

plt.figure(figsize=(8, 6))
plt.plot(nucleos_clavo, tiempo_clavo, marker='o', label='clavo')
plt.title('Tiempo de ejecuci√≥n para "clavo"')
plt.xlabel('N√∫mero de n√∫cleos')
plt.ylabel('Tiempo (segundos)')
plt.grid(True)
plt.legend()
plt.show()

# Gr√°fico para "casero"
nucleos_casero = [dato[0] for dato in datos_casero]
tiempo_casero = [dato[1] for dato in datos_casero]

plt.figure(figsize=(8, 6))
plt.plot(nucleos_casero, tiempo_casero, marker='o', label='casero')
plt.title('Tiempo de ejecuci√≥n para "casero"')
plt.xlabel('N√∫mero de n√∫cleos')
plt.ylabel('Tiempo (segundos)')
plt.grid(True)
plt.legend()
plt.show()
```

Como se puede observar, y como pod√≠amos preveer a medida que incrementamos el n√∫mero de n√∫cleos, disminuye el tiempo de ejecuci√≥n. Sin embargo tambi√©n se obserba un aumento de tiempo al pasar de 4 a 6 n√∫cleos. Esto probablemente se debe a que 4 es mulitplo de 8(bits) y 6 no, por lo tanto el ejercicio es m√°s adecuado para esa cantidad de n√∫cleos.

```python
# Funci√≥n para calcular y/x
def calcular_y_sobre_x(datos):
    valores_y_sobre_x = []
    for dato in datos:
        x, y = dato
        y_sobre_x = y / x
        valores_y_sobre_x.append((x, y_sobre_x))
    return valores_y_sobre_x

# Calcular y/x para cada conjunto de datos
valores_y_sobre_x_hola = calcular_y_sobre_x(datos_hola)
valores_y_sobre_x_clavo = calcular_y_sobre_x(datos_clavo)
valores_y_sobre_x_casero = calcular_y_sobre_x(datos_casero)
```

```python
def crear_grafico(datos, titulo):
    x = [dato[0] for dato in datos]
    y = [dato[1] for dato in datos]
    plt.plot(x, y, marker='o', label=titulo)

# Crear gr√°ficos
plt.figure(figsize=(10, 6))
crear_grafico(valores_y_sobre_x_hola, 'hola')
crear_grafico(valores_y_sobre_x_clavo, 'clavo')
crear_grafico(valores_y_sobre_x_casero, 'casero')
plt.title('Valores de y/x para diferentes palabras')
plt.xlabel('N√∫mero de n√∫cleos')
plt.ylabel('Valor de y/x')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos proporcionados
datos_hola = [2.92, 1.89, 1.90, 1.41]
datos_clavo = [11.55, 9.31, 9.07, 9.45]
datos_casero = [306.67, 260.62, 244.87, 282.99]

# Calcular los nuevos valores
primer_valor_hola = datos_hola[0]
nuevos_datos_hola = [dato / primer_valor_hola for dato in datos_hola]

primer_valor_clavo = datos_clavo[0]
nuevos_datos_clavo = [dato / primer_valor_clavo for dato in datos_clavo]

primer_valor_casero = datos_casero[0]
nuevos_datos_casero = [dato / primer_valor_casero for dato in datos_casero]

# Crear gr√°fico
plt.figure(figsize=(10, 6))

# Graficar datos para 'hola'
plt.plot(range(1, 5), nuevos_datos_hola, marker='o', label='hola')

# Graficar datos para 'clavo'
plt.plot(range(1, 5), nuevos_datos_clavo, marker='o', label='clavo')

# Graficar datos para 'casero'
plt.plot(range(1, 5), nuevos_datos_casero, marker='o', label='casero')

# Configuraci√≥n del gr√°fico
plt.title('Nuevos valores')
plt.xlabel('N√∫mero de n√∫cleos')
plt.ylabel('Nuevos valores')
plt.xticks(range(1, 5))
plt.grid(True)
plt.legend()

# Mostrar gr√°fico
plt.show()
```
