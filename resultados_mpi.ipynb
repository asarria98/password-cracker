# Resultados MPI

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
palabras = ["hola", "clavo", "casero"]

for palabra in palabras:
    # Cargar los datos del archivo CSV sin encabezados
    df = pd.read_csv(f"results/{palabra}_tiempos_mpi.csv", header=None, names=['nucleos', 'tiempo'])

    # Crear el plot
    plt.plot(df.iloc[:, 0], df.iloc[:, 1], label=palabra)

    # Aadir etiquetas y ttulo
    plt.xlabel('Nmero de Ncleos')
    plt.ylabel('Tiempo (s)')
    plt.title(f'Tiempos de ejecucin - {palabra}')
    plt.legend()

    # Mostrar el plot
    plt.show()
```

Como ya vimos en la pruebas realizadas en el ejercicio de paralelizacin de openmp, a medida que incrementamos el nmero de ncleos, disminuye el tiempo de ejecucin. Sin embargo tambin se obserba un aumento de tiempo al pasar de 4 a 6 ncleos. Esto probablemente se debe a que 4 es mulitplo de 8(bits) y 6 no, por lo tanto el ejercicio es ms adecuado para esa cantidad de ncleos.
